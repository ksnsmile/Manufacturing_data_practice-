```python
import pandas as pd

df1=pd.read_csv("C:/Users/sn714/OneDrive/바탕 화면/git/Manufacturing/data/preprocessing_data.csv")

df1.head(3)
```

```python
Y=df1['Target'] #wafer에 불량이라고 판단 된 칩의 개수

X=df1[['Temp_OXid','Oxid_time','thickness','Line_CD','Etching_rate','Temp_Etching']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y,random_state=1234) 
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
y_train_predict=model.predict(X_train)
y_test_predict=model.predict(X_test)
```

```python
r2_score(y_train,y_train_predict)
```

```python
r2_score(y_test,y_test_predict)
```

```python
def evaluation_reg(model):
    y_train_predict=model.predict(X_train)
    y_test_predict=model.predict(X_test)
    print("학습 데이터 결정계수 : ", r2_score(y_train,y_train_predict))
    print("검증 데이터 결정계수 : ", r2_score(y_test,y_test_predict))
    print("학습 데이터 MSE : ", mean_squared_error(y_train,y_train_predict))
    print("검증 데이터 MSE : ", mean_squared_error(y_train,y_train_predict))
```

```python
evaluation_reg(model)
```

# 특성공학 기법 적용

- pipe line : 학습과 특성공학(결측치 제거,변형 ,추가 ) 기법을 동시에 사용하여, 데이터를 한꺼번에 처리하면서 학습을 수행하는 함수

```python
from sklearn.pipeline import Pipeline # 학습 + 특성공학
from sklearn.impute import SimpleImputer #결측값 처리 함수
```

```python
pipe_list=[('imputer',SimpleImputer()),
           ('model', LinearRegression() )]

pipe_model=Pipeline(pipe_list)
```

```python
pipe_model.fit(X_train,y_train)
```

```python
evaluation_reg(pipe_model)
```
